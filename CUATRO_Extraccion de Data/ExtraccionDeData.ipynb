# Metodo loc() y iloc()

#### Son para seleccionar data. 

#### loc()  esta basado en nombres (etiquetas), por lo que tenemos que especificar filas y columnas usando el nombre del index/fila y columna.

#### iloc() esta basado en la posicion numerica entera, por lo que tenemos que especifficar  filas y columnas usando el valor de su posicion entera (posicion empieza por el numero 0) 

#### ejemplo 
#### df.loc[['zeo','two']]
#### df.iloc[[0,2]]
#### df.loc[['zeo' : 'two']]
#### df.iloc[[0 : 2]]

# Selecting with conditions 

####  iloc no puede aceptar una serie booleana, pero si puede aceptar una lista booleana. Tenemos que usar la funcion list() para convertir una Serie a una lista.

```python
# una condicion : seleccionar jugadores con tamaño arriba de 180cm
columns = [1,2,3,4]
df_min.iloc[list(df_min['height_cm'] > 180),columns]
```

```python
# multiples condiciones: 
# seleccionar jugadores con tamaño arriba de 180cm y Argentinos
df_min.iloc[list((df['height_cm']>180)&
                 (df['nationality']=='Argentina')), : ]
```

# Fijar nuevo valor para una celda dentro de un Dataframe 

```python
import pandas as pd 
df = pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df = df[['long_name','age','dob','height_cm','nationality','club']]
df.head()
```

# Fijar valor a una celda 

```python
# actualizar altura de L. Messi
df.loc['L. Messi','height_cm'] = 175
df.iloc[0,3] = 175
# Mostrar Dataframe
df.head()
```

# Fijar valor a toda la columna 

```python
# fijar tamaño de 190 a todos los jugadores
df.loc[:,'height_cm'] = 175
df.head()
```

# Fijar valor a toda la fila 

```python
# obtener todas las columnas que corresponden 
# a jugadores rankeado ultimo en fifa
df.iloc[-1,:]
```

```python
# fijar valor nulo a todas las columnas que corresponden al jugador 
import numpy as np
df.iloc[-1,:] = np.nan
```

```python
# mostrar dataframe
df
```

# Fijar valor a Multiples celdas

```python
# fijar valor a todos los items que hacen match 
# con la lista de nombres
df.loc[['L. Messi','Cristiano Ronaldo'],['height_cm']] = 190
```

```python
# mostrar dataframe
df.head()
```

# Fijar valor para filas que cumplen una condicion 

```python
# fijar valor para filas que cumplen una condicion
columns = ['age','dob','height_cm','weight_kg']
df.loc[df['height_cm'] > 180 , columns] = 0
```

# Filtrar un dataframe con el metodo query() 

```python
import pandas as pd 
df = pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df = df[['long_name','age','dob','height_cm','nationality','club']]
df.head()
```

# query() 

```python
# seleccionar jugadores mayores a 34 años
df.query('age>34')

# ejercicio: escribir el equivalente slicing booleano
df[df['age']>34]
```

```python
# seleccionar jugadores mayores a 34 años de Italia
df.query("age>34 and nationality=='Italy'")
# ejercicio: escribir el equivalente slicing booleano
df[(df['age']>34)&(df['nationality']=='Italy')]
```

```python
# agregar un operador de negacion al primer ejemplo
df.query("not(age>34)")
# ejercicio: escribir el equivalente slicing booleano
df[df['age']!=34]
```

```python
# convertir la altura a metros y seleccionar aquellos que 
# midan mas de 180
df.query("height_cm/100 > 1.8")
# ejercicio: escribir el equivalente slicing booleano
df[df['height_cm']>180]
```

```python
# seleccionar jugadores que nacieron antes de 1990

# revisar el tipo de data
df.dtypes
# convertir el tipo de data de la columna "dob" a datatime
df['dob'] = df['dob'].astype('datetime64[ns]')
df['dob'].dt.year
# query
df.query("dob.dt.year > 1990")
# ejercicio: escribir el equivalente slicing booleano
df.loc[df['dob'].dt.year > 1990,:]
```

# .apply() 


```python
import pandas as pd 
df = pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
df.head()
```

```python
import numpy as np
# usar funcion de numpy y aplicar a la serie
df['age'].apply(np.sqrt)
```

```python
# crear tu propia funcion y aplicar a el dataframe
# kg/m2 = indice de masa corporal
def calcular_imc(row):
    return row['weight_kg']/(row['height_cm']/100)**2
```

```python
df.apply(calcular_imc,axis=1)
```

# Funcion Lambda 

```python
# funcion basica
def sum_values(a,b):
    x=a+b
    return x
```

```python
sum_values(2,3)
```

```python
# funcion lambda (una linea)
sum_values_lambda = lambda a,b:a+b
```

```python
sum_values_lambda(2,3)
```

# Apply + Funcion Lambda 

```python
# usar funcion lambda para convertir serie "height_cm" a metros
df['height_cm'].apply(lambda x:x/100)
# alternativa
df['height_cm']/100
```

```python
# usar funcion lambda para convertir sere "long_name" a mayuscula
df['long_name'].apply(lambda x:x.upper())
# alternativa con atributo str
df['long_name'].str.upper()
```

```python
df['dob']= df['dob'].astype('datetime64[ns]')
# usar funcion lambda para obtener el año de la serie "dob"
df['dob'].apply(lambda x:x.year)
# alternativa con el atributo dt
df['dob'].dt.year
```

```python
# aplicar funcion lambda al dataframe para calcular IMC
df.apply((lambda x:x['weight_kg']/(x['height_cm']/100)**2),axis=1)
```

# Hacer copia a un dataframe con copy() 

## deep = True 

```python
# deep=True por defecto (las modificaciones a la data o indices)
df_copy = df.copy()
```

```python
# actualizar valor en el dataframe original
df.loc['L. Messi','height_cm'] = 180
```

```python
# copia vs dataframe original
df.head(1)
df_copy.head(1)
```

## deep = False 

```python
# deep =False (Cualquier cambio a la data original se refleja 
# en la copia)
df_shallow_copy = df.copy(deep=False)
```

```python
# actualizar valor en el dataframe original
df.loc['Cristiano Ronaldo','height_cm'] = 200
```

```python
# copia vs dataframe original
df.head(2)
df_shallow_copy.head(2)
```
