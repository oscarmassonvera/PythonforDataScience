# Leer .csv de multiples Urls con Pandas 

- https://www.football-data.co.uk/mmz4281/2324/E0.csv
- https://www.football-data.co.uk/mmz4281/2324/E1.csv

#### Link: root + season + league 

```python
# estructura de un link
"https://www.football-data.co.uk/mmz4281/" +"2324" +"/"+"E1"+".csv"
```

```python
# crear una variable root
root = "https://www.football-data.co.uk/mmz4281/"
```

```python
# crear una lista de ligas
leagues = ['E0','E1','E2','E3','EC']
frames = []
# hacer loop en la lista leagues, leer multiples csv y adjuntarlo en a una lista
for league in leagues:
    df = pd.read_csv( root + '2324' + '/' + league +'.csv')
    frames.append(df)    
```

```python
# tamaño de los frames
len(frames)
```

```python
# mostrar 1er,2do,3er elemento
frames[4]
```

# Multiples temporadas 

```python

# crear una lista de ligas
leagues = ['E0','E2','E3']
frames = []
# hacer loop a un rango de temporadas e insertar una nueva columna

for league in leagues:
    for season in range(20,23):
        df = pd.read_csv( root + str(season) + str(season+1) + '/' + league +'.csv')
        df.insert(1,'season',season)
        frames.append(df)    
```

```python
# total de frames (6 temporadas x 3 ligas)
len(frames)
```

```python
# mostrar 1er y ultimo elemento (columna season fue creada)
frames[0]
frames[8]
```

# Organizar data en un diccionario 

```python
# crear un diccionario con nombre original de la liga como llave
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# obtener un elemento del diccionario
dict_countries['Spanish La Liga'] 
```

```python
dict_historical_data = {}
# hacer loop sobre llaves de diccionario
# crear una lista de ligas

# hacer un loop sobre las llaves del diccionario
for league in dict_countries:
    frames = []
# hacer loop por temporadas
    for season in range(20,23):
# se lee el csv q corresponde a la league y optenemos el valor y extrallendo el csv de cada temporada
        df = pd.read_csv( root + str(season) + str(season+1) + '/' 
                         + dict_countries[league] +'.csv')
        df.insert(1,'season',season)
# luego todo eso se almacena el frames
        frames.append(df)   
# Concatenamos las temporadas por liga
    dict_concat = pd.concat(frames)
# Obtenemos el nombre de la liga como llave y despues como valorel dataframe
    dict_historical_data[league] = dict_concat
```
