# Crear una columna basado en 1 condicion: np.where()

```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
```

```python
df_laptops.head(3)
```

```python
import numpy as np
```

```python
# crear un array basado en nivel de precio (price_tier)
np.where(df_laptops['Price_euros'] > 2000,'Caro','Barato')
```

```python
# añadir a nueva columna 
df_laptops['Price_categoria'] = np.where(df_laptops['Price_euros'] > 2000,'Caro','Barato')
```

```python
# mostrar 5 primeras columnas
df_laptops.head()
```

```python
# contar los valores en columna price_tier
df_laptops.value_counts('Price_categoria')
```

# Filtrar un dataframe basado en 2 o mas condiciones 

```python
import pandas as pd
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
df_laptops.head(3)
```

```python
# encontrar laptops Apple que cuesten mas de 2000 euros
(df_laptops['Company'] == "Apple") & (df_laptops['Price_euros'] > 2000)  
```

### & (and) 

```python
df_laptops[(df_laptops['Company'] == "Apple") & 
           (df_laptops['Price_euros'] > 2000)  ]
```

### | (or)

```python
df_laptops['Company'] == 'Apple'
df_laptops['Company'] == 'Dell'
(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')
```

```python
# filtrar dataframe basado en multiples condiciones
df_laptops[(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')].value_counts('Company')
```

```python
# encontrar laptops Apple o Del que cuesten mas de 2000 euros
df_laptops['Company'] == 'Apple'
df_laptops['Company'] == 'Dell'
df_laptops['Price_euros'] > 2000

((df_laptops['Company'] == 'Apple')|(df_laptops['Company'] == 'Dell'))&(df_laptops['Price_euros'] > 2000)
```

```python
# filtrar dataframe basado en multiples condiciones
df_laptops[((df_laptops['Company'] == 'Apple')|(df_laptops['Company'] == 'Dell'))&(df_laptops['Price_euros'] > 2000)].value_counts('Company')
```

# Crear una columna condicional con 2 o mas opciones: np.select() 

```python
import numpy as np
```

```python
# crear un array basado en multiples niveles de precio (+2 opciones)


# crear "conditions" ( condiciones ) y "values" ( valores )
condiciones = [
    
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']>2000)&(df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']>800)&(df_laptops['Price_euros']<=2000),
    df_laptops['Price_euros']<=800,
    
]

valores = ['Muy caro','Caro','Barato','Muy Barato']
# >3000 , # >2000,# >800, <=800
```

```python
# añadir a nueva columna
df_laptops['Precios'] = np.select(condiciones, valores)
```

```python
# mostrar dataframe
df_laptops
```

```python
# contar los valores en columna price_tier
df_laptops.value_counts('Precios')
```

## Ejercicio 

```python
# crear un array basado en multiples tamaños de pantalla ("screen size"+)
# Muy grande >16, Grande>14, Pequeño>12, Muy Pequeño<12

# Crear "conditions" (condiciones) y "Values" (valores)
condiciones = [
    
    df_laptops['Inches']>16,
    (df_laptops['Inches']>14)&(df_laptops['Inches']<=16),
    (df_laptops['Inches']>12)&(df_laptops['Inches']<=14),
    df_laptops['Inches']<=12,
    
]

valores = ['Muy grande','Grande','Pequeño','Muy Pequeño']
# Muy grande >16, Grande>14, Pequeño>12, Muy Pequeño<12
```

```python
# contar los valores en columna Screen_size
df_laptops.value_counts('screen size')
```

# Como encontrar filas duplicadas con duplicated() 

```python
import pandas as pd
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
df_laptops.head(3)
```

## Ejemplo 1 

```python
# encontrar duplicados en 1 columna / serie
df_laptops.duplicated('laptop_ID').value_counts()
```

```python
# mostrar elementos en dataframe con duplicados en columna 'laptop_ID'
df_laptops[df_laptops.duplicated('laptop_ID')]
```

```python
# duplicados en 2 o mas columnas 
duplicated = df_laptops.duplicated(['Product','TypeName','Inches'])
```

```python
# muestra todos los valores duplicados (excepto por el primero - keep='')
df_laptops[duplicated].sort_values(['Product','TypeName'])
```

## Ejemplo 2

#### Encontrar las laptops mas baratas y caras en cada compañia usando el metodo sort_values() y duplicated() (argumentos: keep "first","last" and false) 

```python
# ordenar dataframe ascendente por "company" y "price"
# mas brato primero y mas caro ultimo
df_laptops=df_laptops.sort_values(['Company','Price_euros'])
```

```python
# revisar todos las categorias en la columna "Company"
df_laptops.value_counts('Company')
```

```python
# valores duplicados en la columna "Company"
duplicated_first = df_laptops.duplicated('Company',keep='first')
```

```python
# mostrar dataframe con valores duplicados en la columna "Company"
df_laptops[duplicated_first]
# keep = 'first' (laptops mas baratas por compañia)
# mostrar dataframe con valores no duplicados en la columna "Company"

df_laptops[~duplicated_first][['Company','Price_euros']]

# revisar todas las categorias
df_laptops[~duplicated_first].value_counts('Company')
```

```python
# keep='last'
duplicated_last = df_laptops.duplicated('Company',keep='last')
# keep=False
duplicated_false = df_laptops.duplicated('Company',keep=False)
```

```python
# keep='last'(laptops mas caras por compañia)
# mostrar dataframe con valores no duplicados en columna "Company"
df_laptops[~duplicated_false]
```

```python
# keep=False
# mostrar dataframe con valores no duplicados en columna "Company"
df_laptops.value_counts('Company')
```

#### keep = 'first' ---> mantiene solo el primer valor duplicado
#### keep = 'last' ---> mantiene solo el ultimo valor duplicado
#### keep = False ---> borrar valores duplicados
