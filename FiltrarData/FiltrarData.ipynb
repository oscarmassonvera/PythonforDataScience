# Crear una columna basado en 1 condicion: np.where()

```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
```

```python
df_laptops.head(3)
```

```python
import numpy as np
```

```python
# crear un array basado en nivel de precio (price_tier)
np.where(df_laptops['Price_euros'] > 2000,'Caro','Barato')
```

```python
# a単adir a nueva columna 
df_laptops['Price_categoria'] = np.where(df_laptops['Price_euros'] > 2000,'Caro','Barato')
```

```python
# mostrar 5 primeras columnas
df_laptops.head()
```

```python
# contar los valores en columna price_tier
df_laptops.value_counts('Price_categoria')
```

# Filtrar un dataframe basado en 2 o mas condiciones 

```python
import pandas as pd
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
df_laptops.head(3)
```

```python
# encontrar laptops Apple que cuesten mas de 2000 euros
(df_laptops['Company'] == "Apple") & (df_laptops['Price_euros'] > 2000)  
```

### & (and) 

```python
df_laptops[(df_laptops['Company'] == "Apple") & 
           (df_laptops['Price_euros'] > 2000)  ]
```

### | (or)

```python
df_laptops['Company'] == 'Apple'
df_laptops['Company'] == 'Dell'
(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')
```

```python
# filtrar dataframe basado en multiples condiciones
df_laptops[(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')].value_counts('Company')
```

```python
# encontrar laptops Apple o Del que cuesten mas de 2000 euros
df_laptops['Company'] == 'Apple'
df_laptops['Company'] == 'Dell'
df_laptops['Price_euros'] > 2000

((df_laptops['Company'] == 'Apple')|(df_laptops['Company'] == 'Dell'))&(df_laptops['Price_euros'] > 2000)
```

```python
# filtrar dataframe basado en multiples condiciones
df_laptops[((df_laptops['Company'] == 'Apple')|(df_laptops['Company'] == 'Dell'))&(df_laptops['Price_euros'] > 2000)].value_counts('Company')
```
