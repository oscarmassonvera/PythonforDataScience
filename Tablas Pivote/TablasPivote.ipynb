# Introduccion a tablas pivote en pandas

## pivot() : reconstruye una data basada en valores de una columna (no soporta operaciones)
## pivot_table() : crea una tabla pivote al estilo de excel (soporta operaciones)

```python
import pandas as pd
```

## pivot() 

```python
# read and show the dataset
df_gdp = pd.read_csv('gdp.csv',encoding='latin-1')
df_gdp.head()
```

```python
# reshape the dataframe with .pivot()
df_gdp.pivot(index='year',columns='country',values='gdppc').head()
```

## .pivot_table()

```python
# leer y mostrar el dataset
df_sales = pd.read_excel('supermarket_sales.xlsx')
df_sales.head()
```

```python
# hacer una tabla pivote y sumar valores agrupados
df_sales.pivot_table(index='Gender', 
                     aggfunc="sum",
                     values=['Quantity','Total'])
```

```python
# hacer una tabla pivote que diga cuando mujeres y 
# hombres gastaron y en q categorias
df_sales.pivot_table(index='Gender', 
                     aggfunc="sum",
                     values='Total',
                     columns='Product line')
```
